# Introduction

NB:

- API subject to change
- Oanda has two versions of REST API - v1 and v20
- Newer version is the v20 REST API
- However, there are still a lot of accounts that uses v1 REST API
- v1 account number format is 7 digits long
- v20 account number format is in the form of 000-000-0000000-000
